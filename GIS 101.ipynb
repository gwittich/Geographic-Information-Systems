## **Vector Layers**

*   **Point** (red) - Corodintes
*   **Lines** (black) - Represent the LENGTH of things, Connect one or more **Points**
*   **Polygons** (green) - Represent the AREA of things, Made up of 3 or more non-intersecting **Lines**

Vector data contaions attributes (text or numerical) stored as fields in a table









1) Vector layers offer the ability to associate attributes or a database of text in numerical fields against the features on the map.

These attribute values can be queried to quickly select, manage, and map subsets of vector data.

eg. You may want to have a data set of rare flora occurrences across a region and need a map to show only orchid species found in the last five years. if each feature records a species name than a date of observation the standard query tools in a GIS can be used to quickly select and only show these on your map with a different color for each species.

2) The special relationship between features is also a major advantage of vector layers a GIS.

eg. You may need to know how many landowners have their properties adjacent to a river the resulting data set attributes could then tell you who those adjacent landowners are and how to get in contact with them.

## **Coordinate Reference Systems**

defines where data is located in the spatial world.

```python
Image('Screen Shot 2022-07-21 at 12.36.25 pm.png')
```

**Geographic CRS**

*   Latitude/Longitude
*   **Units**: Degrees, minutes, seconds or decimal degrees
*   Based on spherical model
*   Unique coordinates anywhere on Earth

**Projected CRS**

*   Eastings and Northings
*   **Units**: Metres
*   Projects parts of the Earth onto a *flat plane*
*   Same coordinates in multiple zones



```python
Image('Screen Shot 2022-07-21 at 12.36.48 pm.png')
```

**Geographic CRS**

*   Example: GDA94
*   Longitude:	 118.94332
*   Latitude:	-34.11793 

**Projected CRS**

*   Example: MGA94 Zone 50
*   Easting:	679225mE
*   Northing:	6223062mN

A **Geographic CRS** is based on approximating the earth as a **sphere** and represents coordinates as **latitudes and longitude** calculating areas accurately.
There are **many Geographic CRS**

A **Projected CRS** attempts to project parts of the Earth's surface onto a **flat plane** and coordinates for projected systems come in the form of **eastings and northings**.

There are **many Projected CRS**. Each is stronger in one area or another, eg. some Projected CRS are better for calculating **areas** accurately, others better for 
calculating **distances** accurately and still others are better for visually accurate displays. 

There are many projected CRS is out there and each one tries to get one of these aspects correct.

To show spatial data correctly on a map you need to know:
*   the **coordinates** (location values) **and** 
*   the **Coordinate Reference System** used.

The coordinates of a single location are very different when expressed in different Coordinate Reference Systems.

## **Metadata**

GIS layers have files coerresponding descriptions of the data - Metadata.
*   Who created it ?
*   Why was it created?
*   When was it published?
*   What does it contain?
*   Is it complete?
*   Who should I speak to for more information?

Metadata can be very important because it helps us to **understand if the data is suitable for our particular mapping or GIS analysis tasks**.
It helps to avoid inappropriate use of the data where it's not obvious from looking at the data itself.
EG. Using layers that are **out of date** is a common trap where you're attempting to map the current state of something.

Metadat can be in the form of and text doc, word doc, xml doc.



There are **ISO standards relating to Geographic information metadata** which are designed to bring consistency and ensure quality in the way that we describe spatial information. The most notable of these is ISO:19115
https://www.iso.org/standard/53798.html
